## Beginning thoughts
It is our goal, to demonstrate the connection between the pitch of a sound and its frequency. In order to do that, a function to perform a discrete fourier transform needs to be coded. The first question arising would be the choice of a fitting file format for our sound file. We decided to use the Waveform Audio File Format (.wav) because of the availability of software able to record in this format and the fact that the data is saved uncompressed making sure no information is lost which is not true for the other popular file format MP3 (.mp3).
In order to read the .wav-file the function *wavfile.read* from the Python library *scipy.io* is used. *wavfile.read* can take two inputs, the first one being the filename as a string eg. 'Testfile.wav' or an open file handle and the second optional one which declares whether the data should be read as memory mapped as is set to FALSE by default. *wavfile.read* outputs two elements. The first one is the sampling rate used for the recording and the second one is an array containing the amplitude of the sound during the recorded time frame.
The library *numpy* is used for mathematical functions and values such as the sine function or the number $\pi$.
*matplotlib.pyplot* is a library containing functions to generate plots.
*dft_wav* contains the functions we coded for performing the discrete fourier transform and the inverted discrete fourier transform.

```python
import numpy as np
import matplotlib.pyplot as p
from scipy.io import wavfile
import dft_wav as dft
```

# Testing
The functions performing the transformations are located in the *dft_wav.py* file. Comments regarding the coding can be read there. So let's begin the testing of our functions. First we should generate a signal with known frequency $\omega$ and see whether the DFT is working. 

```python
N = 2000
x = np.linspace(0,20*np.pi,N) #creating a discrete number of time points

test_signal = np.sin(10*x) #generating a sine signal
test_rate=N/(20*np.pi) #rate used for the time points 
p.xlabel('time [s]')
p.ylabel('amplitude [a.u.]')
p.plot(x[0:100],test_signal[0:100]) #plotting parts of the sound signal
```

```python
DFT_x, DFT_signal=dft.dft(test_signal,test_rate) #performing the DFT
p.xlabel('frequency [Hz]')
p.ylabel('amplitude [a.u.]')
p.plot(DFT_x,np.abs(DFT_signal)) #ploting the fourier transform 
```

```python
p.xlabel('frequency [Hz]')
p.ylabel('amplitude [a.u.]')
p.plot(DFT_x[50:150],np.abs(DFT_signal)[50:150]) #ploting the fourier transform
```

The DFT is working however there is a second peak visible at the maximum frequency minus the expected frequency. This is a normal effect because of the discretization. As long as you can be sure that your sound signal does not contain any frequency larger than half the sample rate of the recording, every peak in frequency above half the sample rate can be ignored. If this assumption is not true the effect of aliasing occurs. Higher frequencies above half the sample rate cause a peak below it so that the result can not be trusted anymore.

The inverse DFT is working as well. Now let's test the core properties of the DFT:

```python
def linearity_test(signal1,signal2,rate1=1,rate2=1):
    #testing the linearity of the DFT for test signals with same sample rate
    if type(signal1) == str:
        assert signal1.endswith(".wav"), "file is not a wavefile"
        rate1, v1 = wavfile.read(signal1)
    else:
        v1 = signal1
    assert len(np.shape(v1)) == 1, "input array needs to be one-dimensional"
    if type(signal2) == str:
        assert signal2.endswith(".wav"), "file is not a wavefile"
        rate2, v2 = wavfile.read(signal2)
    else:
        v2 = signal2
    assert len(np.shape(v2)) == 1, "input array needs to be one-dimensional"
    
    if rate1!=rate2:
        raise linearity_test('Your sample rates must be equal!')
    
    if len(v1)>11000: #limiting the test file to prevent crash because of ram usage
        v1=v1[0:11000]
    if len(v2)>11000:
        v2=v2[0:11000]
    alpha=10*(np.random.rand()-0.5)+1j*10*(np.random.rand()-0.5)
    beta=10*(np.random.rand()-0.5)+1j*10*(np.random.rand()-0.5)
   
    v1=v1[0:min(len(v1),len(v2))]
    v2=v2[0:min(len(v1),len(v2))]
    xt1,vt1=dft.dft(v1,rate1)
    xt2,vt2=dft.dft(v2,rate2)
    
    sum_v=alpha*v1+beta*v2
    sum_xt, sum_vt=dft.dft(sum_v,rate1)
    N=len(v1)
    diff=np.abs(sum_vt[0:N]-(alpha*vt1+beta*vt2)[0:N])
    rel_diff=diff/np.abs(sum_vt[0:N])
    
    if max(rel_diff)<1e-10: #making sure the maximum value of the relative difference is extremely low
        return "Linearity is given to a relative precision of 1e-10"
    else:
        return "Linearity is not given"
```

The inversion formula should be fullfilled as well:

```python
def inversion_test(signal,rate=1):
    #testing the inversion of the DFT 
    if type(signal) == str:
        assert signal.endswith(".wav"), "file is not a wavefile"
        rate, v = wavfile.read(signal)
    else:
        v = signal
    assert len(np.shape(v)) == 1, "input array needs to be one-dimensional"
    
    if len(v)>11000:
        v=v[0:11000]
    
    xt,vt=dft.dft(v,rate)
    vtt=dft.idft(vt)
    N=len(vt)
    diff=np.abs(vtt[0:N]-v[0:N])
    
    if max(diff)<1e-10:
        return "Inversion formula is fullfilled to a precision of 1e-10"
    else:
        return "Inversion formula is not fullfilled"
```

The identity for repeated DFT with shifted values is tested:

```python
def inversion2_test(signal,rate=1):
    #testing the inversion of the DFT 
    if type(signal) == str:
        assert signal.endswith(".wav"), "file is not a wavefile"
        rate, v = wavfile.read(signal)
    else:
        v = signal
    assert len(np.shape(v)) == 1, "input array needs to be one-dimensional"
    
    if len(v)>11000:
        v=v[0:11000]
    
    xt,vt=dft.dft(v,rate)
    xtt,vtt=dft.dft(vt,rate)
    N=len(vt)
    vtt=vtt[::-1]
    diff=np.abs(1/N*vtt[0:N-1]-v[1:N])
    
    if max(diff)<1e-10:
        return "Inversion formula is fullfilled to a precision of 1e-10"
    else:
        return "Inversion formula is not fullfilled"
```

And last but not least the generalized parseval identity can be tested:

```python
def parseval_test(signal1,signal2,rate1=1,rate2=1):
    #testing the generalized parseval identity of the DFT for test signals with same sample rate
    if type(signal1) == str:
        assert signal1.endswith(".wav"), "file is not a wavefile"
        rate1, v1 = wavfile.read(signal1)
    else:
        v1 = signal1
    assert len(np.shape(v1)) == 1, "input array needs to be one-dimensional"
    if type(signal2) == str:
        assert signal2.endswith(".wav"), "file is not a wavefile"
        rate2, v2 = wavfile.read(signal2)
    else:
        v2 = signal2
    assert len(np.shape(v2)) == 1, "input array needs to be one-dimensional"
    
    if rate1!=rate2:
        raise linearity_test('Your sample rates must be equal!')
    
    if len(v1)>11000:
        v1=v1[0:11000]
    if len(v2)>11000:
        v2=v2[0:11000]
    v1=v1[0:min(len(v1),len(v2))]
    v2=v2[0:min(len(v1),len(v2))]
      
    v1=v1/max(abs(v1)) #linearity is proven rescaling for easier calculation
    v2=v2/max(abs(v2))
    
    
    xt1,vt1=dft.dft(v1,rate1)
    xt2,vt2=dft.dft(v2,rate2)
    
  
    
    N=len(v1)
    diff=np.abs((np.vdot(v1,v2))-1/N*np.vdot(vt1,vt2))
    rel_diff=diff/np.abs(np.vdot(v1[0:N],v2[0:N]))
 
    if rel_diff<1e-10:
        return "Generalized Parseval Identity is fullfilled to a relative precision of 1e-10"
    else:
        return "Generalized Parseval Identity is not fullfilled"
```
